```python
from PIL import Image
```

```python
import pytesseract as pt
```

```python
def simple_ocr(*imgs):
  for img in imgs:
    text = (pt.image_to_string(Image.open(img)))
    with open("text.txt", 'a') as f:
       print((text), file = f)
```

```python
simple_ocr('/content/wk_1.png', '/content/wk_2.jpg', '/content/wk_3.png')
```
